[Question](https://leetcode.com/problems/populating-next-right-pointers-in-each-node/description/)

Given a binary tree

```
struct TreeLinkNode {
  TreeLinkNode *left;
  TreeLinkNode *right;
  TreeLinkNode *next;
}
```

Populate each next pointer to point to its next right node. If there is no next right node, the next pointer should be set to NULL.

Initially, all next pointers are set to NULL.

Note:

- You may only use constant extra space.
- Recursive approach is fine, implicit stack space does not count as extra space for this problem.
- You may assume that it is a perfect binary tree (ie, all leaves are at the same level, and every parent has two children).


Example:

Given the following perfect binary tree,
```
     1
   /  \
  2    3
 / \  / \
4  5  6  7
```

After calling your function, the tree should look like:

```
     1 -> NULL
   /  \
  2 -> 3 -> NULL
 / \  / \
4->5->6->7 -> NULL
```

#### Level Order Traversal

````python
# Definition for binary tree with next pointer.
class TreeLinkNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None
        self.next = None


class Solution:
    # @param root, a tree link node
    # @return nothing
    def connect(self, root):
        if root is None:
            return
        queue = []
        queue.append((root, 0))
        previous_depth = -1
        previous_node = None
        while queue:
            root, depth = queue.pop(0)
            if not root:
                continue
            if previous_depth == depth:
                previous_node.next = root
                previous_node = root
            else:
                previous_node = root
                previous_depth = depth
            queue.append((root.left, depth + 1))
            queue.append((root.right, depth + 1))
````

#### Recursive
**Note:** This approach works only for **_Complete Tree_**

We will be traversing like Preorder traversal and connect all the parents first. So when we will come to it's child node 2 cases arises:

**Case 1**

Child node is left child -> It's connection will be with right child of it's parent

**Case 2**

Child node is right child -> It's connection will be with left child of next node of it's parent

````python
# Definition for binary tree with next pointer.
class TreeLinkNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None
        self.next = None

class Solution:
    # @param root, a tree link node
    # @return nothing
    def connect(self, root):
        if not root:
            return
        self.connect_preorder_traversal(root, None, False)

    def connect_preorder_traversal(self, root, parent, is_left):
        if not root:
            return
        if parent:
            if is_left:
                root.next = parent.right
            else:
                if parent.next:
                    root.next = parent.next.left
        self.connect_preorder_traversal(root.left, root, True)
        self.connect_preorder_traversal(root.right, root, False)
````
